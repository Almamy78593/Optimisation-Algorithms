***Résolution du TSP avec un essaim particulaire (PSO)***

```python
tsp_pso_9_villes = tspPso(f=costPso, distances=DISTANCES_VILLES, n_particules=N_PARTICULE, n_iterations=ITERATION_MAX_TSP)
print("Meilleur particule retenu:",tsp_pso_9_villes[0], " : ", tsp_pso_9_villes[1][-1], "\nNb_iteration : ",tsp_pso_9_villes[3])
print("Arret progression vers la meilleur solution :", tsp_pso_9_villes[4])
plotResolvingTsp(tsp_pso_9_villes)

nb_essai = 0
while tsp_pso_9_villes[1][-1] > SOLUTION_TSP : 
  tsp_pso_9_villes = tspPso(f=costPso, distances=DISTANCES_VILLES, n_particules=N_PARTICULE, n_iterations=ITERATION_MAX_TSP)
  nb_essai+=1

print("Meilleur particule retenu:",tsp_pso_9_villes[0], " : ", tsp_pso_9_villes[1][-1], "\nNb_iteration : ",tsp_pso_9_villes[3])
print("Arret progression vers la meilleur solution :", tsp_pso_9_villes[4])
print("Nombre de fois que l'on éxecute PSO pour trouver la meilleure solution : ", nb_essai)

plotResolvingTsp(tsp_pso_9_villes)
```
